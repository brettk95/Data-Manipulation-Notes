```python
import pandas as pd
import numpy as np

df = pd.read_csv("athlete_events.csv")
df.head()
```

### Specifying a single column to fill in missing values for

```python
# one line code to replace missing values in Height by the mean height of group - where the group by is by Sport

df["Height"] = df.groupby("Sport")["Height"].transform(lambda x: x.fillna(x.mean()))
df["Height"]
```

```python
# notice that there are still some missing values in the Height column because some sports had missing heights for all athletes
print(df["Height"].isnull().any())

# these rows are:
df[df["Height"].isnull()].head()
```

```python
# so need another piece of code to fill in these missing values with the mean of the mean heights groupbed by sport

df["Height"] = df["Height"].fillna(df.groupby("Sport")["Height"].mean().mean())
```

```python
# I could have also used the mean of the Height column, without taking into account the groups:

df["Height"] = df["Height"].fillna(df["Height"].mean())
```

### Generalize fill missing to multiple columns

```python
df[["Height","Weight"]] = df.groupby("Sport")[["Height","Weight"]].transform(lambda x: x.fillna(x.mean()))
```

```python
df[["Height","Weight"]] = df[["Height","Weight"]].apply(lambda x: x.fillna(x.mean()))
```

```python
print(df["Height"].isna().any(), df["Weight"].isna().any())
```
